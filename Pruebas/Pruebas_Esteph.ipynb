```python
# Importar las librerías necesarias
import pandas as pd
import re
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../Datos/train.csv")                                   
df['Text'] = df['Title'] + " " + df['Opinion']
df = df.dropna().reset_index(drop=True)                                        # Eliminar filas con nan's 
```

```python
# Definir X e y
X = df['Text']  # El texto combinado que se usará como entrada
y = df['Label']  # Las etiquetas que se usarán para la clasificación

# Dividir el conjunto de datos en entrenamiento y validación, estratificado por y
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```
